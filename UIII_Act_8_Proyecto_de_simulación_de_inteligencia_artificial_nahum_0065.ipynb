```python
# =====================================================
# SIMULADOR DE IA QUÍMICA - PROBETA INTELIGENTE
# Nivel: Principiante
# =====================================================

print("========================================")
print("   LABORATORIO IA - PROBETA INTELIGENTE ")
print("========================================")

# Variables iniciales
energia = 100
mezcla = ""
temperatura = 25
opcion = ""

# Ciclo principal
while opcion != "5":

    print("\n========= MENÚ PRINCIPAL =========")
    print("1. Agregar sustancia")
    print("2. Calentar probeta")
    print("3. Analizar mezcla")
    print("4. Ver estado del laboratorio")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCIÓN 1
    if opcion == "1":
        print("\n--- SUSTANCIAS DISPONIBLES ---")
        print("a) Ácido")
        print("b) Agua")
        print("c) Metal")

        sustancia = input("Elige una sustancia: ")

        if sustancia == "a":
            mezcla = "Ácido"
            energia = energia - 10
            print("La IA agregó Ácido a la probeta.")

        elif sustancia == "b":
            mezcla = "Agua"
            energia = energia - 5
            print("La IA agregó Agua a la probeta.")

        elif sustancia == "c":
            mezcla = "Metal"
            energia = energia - 15
            print("La IA agregó Metal a la probeta.")

        else:
            print("Sustancia no válida.")

    # OPCIÓN 2
    elif opcion == "2":

        if mezcla == "":
            print("Primero debes agregar una sustancia.")

        else:
            print("\nCalentando probeta...")
            temperatura = temperatura + 20
            energia = energia - 20

            print("Temperatura actual:", temperatura, "grados")

            if temperatura >= 80:
                print("⚠ ALERTA: La mezcla está muy caliente.")

            else:
                print("La temperatura es segura.")

    # OPCIÓN 3
    elif opcion == "3":

        print("\n--- ANALISIS IA ---")

        if mezcla == "Ácido":
            print("Resultado: Sustancia corrosiva detectada.")

        elif mezcla == "Agua":
            print("Resultado: Sustancia estable y segura.")

        elif mezcla == "Metal":
            print("Resultado: Material conductor encontrado.")

        else:
            print("La probeta está vacía.")

    # OPCIÓN 4
    elif opcion == "4":

        print("\n===== ESTADO DEL LABORATORIO =====")
        print("Energía:", energia)
        print("Temperatura:", temperatura)
        print("Contenido de la probeta:", mezcla)

        if energia <= 20:
            print("⚠ La energía del laboratorio es baja.")

        else:
            print("El laboratorio funciona correctamente.")

    # OPCIÓN 5
    elif opcion == "5":
        print("\nApagando sistema de IA química...")
        print("Gracias por usar el laboratorio inteligente.")

    # OPCIÓN INVÁLIDA
    else:
        print("Opción incorrecta. Intenta nuevamente.")

# Mensaje final
print("\n====================================")
print("Programa realizado por:")
print("Nahum Flores NL = 21")
print("====================================")
```
